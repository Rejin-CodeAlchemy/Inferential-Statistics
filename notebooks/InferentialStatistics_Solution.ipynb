A physiotherapist with a male football team is interested in studying the relationship between foot injuries and the positions at which the players play from the data collected.

| Position | Striker| Forward | Attacking Midfielder | Winger | Total |

| Players Injured     | 45 | 56 | 24 | 20 | 145 |
| Players Not Injured | 32 | 38 | 11 |  9 |  90 |

Based on the above data, answer the following questions.

```python
import numpy as np
import pandas as pd
```

#### 1.1 What is the probability that a randomly chosen player would suffer an injury?

```python
# Total number of players that have suffered injuires = 145
# Total number of players in the sample set = 235
# Let Pa be the probability that a random player be chosen that would have suffer an injury.
P= np.round(145/235,2)
print('Probability that a randomly chosen player would suffer an injury is ',P) 
```

#### 1.2 What is the probability that a player is a forward or a winger?

```python
# Let Pa be the probability of a player chosen is Forward.
# And Pb be the probability of a player chosen is winger.
# As one player can not be played as both Forward and Winger, this is clearly a Mutually Exclusive sets.
# Then probability that a player is a forward or a winger is Pa U Pb.

# Pa U Pb = Pa + Pb - Pa n Pb (Pa n Pb =0 for mutually exclusive events)
# Pa U Pb = Pa + Pb
Pa = np.round(94/235,2)
Pb = np.round(29/235,2)

print('Probability that a player is a forward or a winger is ', Pa + Pb) 
```

#### 1.3 What is the probability that a randomly chosen player plays in a striker position and has a foot injury?

```python
# Let Pa be the probability that a randomly chosen player plays in a striker position and has a foot injury.
# This is an example of Joint Probability

Pa = np.round(45/235,2)  

print('Probability that a randomly chosen player plays in a striker position and has a foot injury is ', Pa)
```

#### 1.4 What is the probability that a randomly chosen injured player is a striker?

```python
# Here getting a striker from the set of injured persons.
Pa = np.round(45/145,2)

print('Probability that a randomly chosen injured player is a striker is ', Pa)
```

The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; 
##### (Provide an appropriate visual representation of your answers, without which marks will be deducted)


```python
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats 
from scipy.stats import norm
```

#### 2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

```python

mu = 5 #mean
sigma = 1.5 #standard deviation
# cumulative probability 

prob_less_than_317 = norm.cdf(3.17, mu, sigma)

print('Proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm is ',np.round(prob_less_than_317,2))
```

```python
x = np.linspace(0,10, 50)
plt.plot(x, norm.pdf(x, mu, sigma)) # plot the pdf of the normal distribution
plt.axvline(x=3.17, c="r") # draw a red vertical line at x = 3.17

x1 = np.linspace(0, 3.17, 50) # create an array of 50 numbers between 0 and 3.17
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color="r") # fill the specified region with red color
plt.xlabel('''Breaking Strength (in Kg per sq cm)

Figure 2.1''') # set the x-axis label
plt.ylabel("Proportion") # set the y-axis label
plt.title("Breaking strength < 3.17 kg/sq cm") # set the title
plt.show() # display the plot
```

```python

alloy_1 = df1[df1['Alloy'] == 1]
alloy_2 = df1[df1['Alloy'] == 2]
```

Null Hypothesis is :

>$H_0: $The hardness of implants does not vary depending on dentists.

Alternate Hypothesis Ha: 

>$H_a: $ The hardness of implants varies depending on dentists.

```python
# Function to check assumptions of ANOVA
# 1. Normality: Shapiro-Wilk test
# 2. Homogeneity of variances: Levene's test

def check_assumptions(data, group_col, response_col):
    # Normality test
    groups = data[group_col].unique()
    normality_p_values = {}
    for group in groups:
        stat, p_value = stats.shapiro(data[data[group_col] == group][response_col])
        normality_p_values[group] = p_value
    
    # Homogeneity of variances test
    stat, levene_p_value = stats.levene(*[data[data[group_col] == group][response_col] for group in groups])
    
    return normality_p_values, levene_p_value

# Check assumptions for Alloy 1
normality_p_values_alloy1, levene_p_value_alloy1 = check_assumptions(alloy_1, 'Dentist', 'Response')

print('Alloy 1 Normality p-values:', normality_p_values_alloy1)
print('Alloy 1 Levene p-value:', levene_p_value_alloy1)
```

* For Alloy 1, the Normality p-values for all dentists are greater than 0.05. So the data is normally distributed.
* The Levene's test p-value for Alloy 1 is greater than 0.05. So the variances are homogeneous.

```python
# Check assumptions for Alloy 2
normality_p_values_alloy2, levene_p_value_alloy2 = check_assumptions(alloy_2, 'Dentist', 'Response')
print('Alloy 2 Normality p-values:', normality_p_values_alloy2)
print('Alloy 2 Levene p-value:', levene_p_value_alloy2)
```

* For Alloy 2, the normality p-values for all dentists except 4 are greater than 0.05. Hence the data is mostly normally distributed.
* The Levene's test p-value for Alloy 2 is greater than 0.05, indicating that the variances are homogeneous.

* Since the assumptions are mostly satisfied,we can use ANOVA test to determine if there are significant differences in implant hardness depending on dentists.

```python
# ANOVA test
anova_alloy1_fstat, anova_alloy1_p_value = stats.f_oneway(*[alloy_1[alloy_1['Dentist'] == dentist]['Response'] for dentist in alloy_1['Dentist'].unique()])
print('ANOVA p-value for Alloy 1:', anova_alloy1_p_value)
```

```python
anova_alloy2_fstat, anova_alloy2_p_value  = stats.f_oneway(*[alloy_2[alloy_2['Dentist'] == dentist]['Response'] for dentist in alloy_2['Dentist'].unique()])
print('ANOVA p-value for Alloy 2:', anova_alloy2_p_value)
```

* For both Alloy 1 and Alloy 2, the p-value is greater than 0.05. 
* This means we fail to reject the null hypothesis, indicating that the hardness of implants does not vary depending on dentists.
